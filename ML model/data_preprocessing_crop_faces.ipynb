```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
fpath = glob('./data/female/*.jpg')
mpath = glob('./data/male/*.jpg')
```

```python
print("No. of female photos", len(fpath))
print("No. of male photos", len(mpath))
```

```python
# Read image and convert it to rgb
for i in range(len(mpath)):
    try:
        img = cv2.imread(mpath[i]) #read image in bgr always
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # convert bgr to rgb
        
        # Apply haar cascade classifier
        
        haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
        gray = cv2.cvtColor(img_rgb,cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray,1.5,5)
        for x,y,w,h in faces_list:
            # cv2.rectangle(img_rgb,(x,y),(x+w,y+h),(0,255,0),2)
            # crop face
            roi = img[y:y+h,x:x+w]
            cv2.imwrite(f'./crop_data/male/male_{i}.jpg',roi)
            print('Image Processed Succesfully')
    except:
        print('Not Processed')
```

```python
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```
